```python
import torch
from torchviz import make_dot
from models.Transformer_model import AgentPredictor
from easydict import EasyDict

# Example configuration dictionary
config = {
    'map_input_size': 3,
    'map_hidden_size': 256,
    'agent_his_input_dim': 5,
    'agent_his_hidden_size': 256,
    'fusion_hidden_size': 256,
    'num_fusion_layer': 2,
    'dropout_rate': 0.1,
    'in_hidden_dim': 256,
    'head_hidden_dim': 256,
    'out_traj_dim': 40 * 5,
    'num_queries': 40,
    'num_decoder_layer': 2,
    'aux_loss_temponet': True,
    'aux_loss_spanet': True,
    'lr': 1e-3
}

config = EasyDict(config)

# Initialize model
model = AgentPredictor(config)

# Dummy input
batch_size = 1
seq_length = 10
map_length = 50
dummy_input = {
    'ego_stat': torch.randn(batch_size, seq_length, 12),
    'ego_map': torch.randn(batch_size, map_length, 3),
    'ego_fut': torch.randn(batch_size, 40, 12),
    'ahead_stat': torch.randn(batch_size, seq_length, 5),
    'ahead_fut': torch.randn(batch_size, 40, 5),
    'right_stat': torch.randn(batch_size, seq_length, 5),
    'right_fut': torch.randn(batch_size, 40, 5),
    'ego_map_mask' : torch.ones((batch_size, 50))
}

# Forward pass
output, aux_temponet_pred = model(dummy_input)

# Select 'agent' tensor for visualization
main_output_tensor = output['pos']

# Draw computation graph
graph = make_dot(main_output_tensor, params=dict(model.named_parameters()))

# Display graph in Jupyter Notebook
graph.format = 'png'
graph.render("agent_predictor_architecture")
from IPython.display import Image
Image(filename="agent_predictor_architecture.png")
```
